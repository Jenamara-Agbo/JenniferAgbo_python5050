```python
#Yout code here:
rand = pd.read_csv("RAND_Database_of_Worldwide_Terrorism_Incidents.csv", encoding="ISO-8859-1")
rand.head()
```

```python
# Add a new column 'Case_ID' starting from 1 for easier identification of entrie
rand.insert(0, 'Case_ID', range(1, len(rand) + 1))

# Set the 'Serial' column as the index
rand.set_index('Case_ID', inplace=True)

# Display the DataFrame to confirm changes
rand.head()
```

```python
#for a Text-Based Keyword Matching
# Define keywords for each category.
# Check each incident description for these keywords.
# Assign a motivation category based on matches.

# Expanded keyword lists for each category
economic_keywords = [
    "ransom", "extortion", "payment", "theft", "looting", "bank", "oil", 
    "resource", "economy", "trade", "money", "cash", "bribe" ,"demanded money", "financial gain", "economic sanctions",
    "funding", "wealth", "smuggling", "drug trafficking", "financial"
]

political_keywords = [
    "liberation", "freedom", "revolution", "independence", "resistance", 
    "government", "governance", "state", "regime", "political", "sovereignty", "national",
    "separatist", "self-determination", "dictatorship", "republic", "anti-government", 
    "democracy", "authority", "power", "injustice", "reform", "uprising", "conflict",
    "rights", "ideology", "opposition", "military", "occupation", "movement"
]

religious_keywords = [
    "jihad", "holy war", "sacred", "faith", "religious beliefs", "god", "divine", 
    "islamic", "christian", "hindu", "christianity" "jewish", "buddhist", "martyr", "holy", 
    "worship", "believer", "crusade", "infidel", "apostate", "evangelism", "heaven",
    "religion", "blasphemy", "prophet", "convert", "sect", "sharia", "fatwa", "imam", "pastor"
]

# Function to categorize motivations
def categorize_motivation(description):
    # Convert description to lowercase for case-insensitive matching
    description = str(description).lower()

    # Initialize flags - using list comprehension
    economic_flag = any(keyword in description for keyword in economic_keywords)
    political_flag = any(keyword in description for keyword in political_keywords)
    religious_flag = any(keyword in description for keyword in religious_keywords)
    
    # Determine category based on flags
    if economic_flag and not (political_flag or religious_flag):
        return "Economic"
    elif political_flag and not (economic_flag or religious_flag):
        return "Political"
    elif religious_flag and not (economic_flag or political_flag):
        return "Religious"
    elif economic_flag and political_flag and not religious_flag:
        return "Economic-Political"
    elif political_flag and religious_flag and not economic_flag:
        return "Political-Religious"
    elif economic_flag and religious_flag and not political_flag:
        return "Economic-Religious"
    elif economic_flag and political_flag and religious_flag:
        return "Economic-Political-Religious"
    else:
        return "Uncategorized"

# Apply categorization function to the 'Description' column
rand['Motivation'] = rand['Description'].apply(categorize_motivation)

# Display a sample of the dataset with the new 'Motivation' column
print(rand[['Description', 'Motivation']].head(10))

print("\n=================================================================================================\n")


# Count the occurrences of each motivation category
motivation_counts = rand['Motivation'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
motivation_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding title and labels
plt.title("Frequency of Terrorism Motivations", fontsize=16)
plt.xlabel("Motivation Category", fontsize=14)
plt.ylabel("Number of Incidents", fontsize=14)

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
